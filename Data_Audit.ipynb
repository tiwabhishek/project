```python
Variable_type=[]
for n in df.columns.tolist():
    if n == 'RainTomorrow':
        Variable_type.append('Target')
    else:
        Variable_type.append('Feature')


print(Variable_type)
Data_audit["Variable's type"] = Variable_type
```

```python
Data_audit["Percentage of missing values"]=((df.isnull().sum() / len(df)) * 100).values.round(1)
```

```python
Categorical_Quantitative =[]
for n in df.dtypes.tolist():
    if n == object:
        Categorical_Quantitative.append('Categorical')
    else:
        Categorical_Quantitative.append('Quantitative')


print(Categorical_Quantitative)
Categorical_Quantitative[0]="not sure"
Data_audit["Categorical/Quantitative"] = Categorical_Quantitative
```

```python
unique_values_in_objt_coloumns=[]
for n in df.columns:
    if df[n].dtype==object:
        unique_values_in_objt_coloumns.append(df[n].nunique())
print(unique_values_in_objt_coloumns)
```

The template does say any thing about object columns with more then 10 values so i just took out all top 10 values

```python
unique_values=[]
for n in df.columns:
    if df[n].dtype==object:
        if df[n].nunique()>10:
            unique_values.append(df[n].value_counts().index[:10].tolist())
        else:
            unique_values.append(df[n].unique())
    else:
        unique_values.append(df[n].describe().round(1))
unique_values[0]="about date i m not sure"   
print(unique_values)
```
